```python
import matplotlib.pyplot as plt
```

```python
# Plots for problem 1 


# Data
categories = ['Total', 'Recursive', 'Reverse']
fully_optimized = [0.21, 0.04, 0.04]  # Fully optimization times
half_optimized = [0.33, 0.08, 0.04]  # Half optimization times
no_optimization = [3.58, 1.12, 2.33]  # No optimization times

# Create figure and axis
fig, ax = plt.subplots()

# Set the position of the bars on the x-axis
x = range(len(categories))

# Make the plot
ax.bar(x, fully_optimized, width=0.2, label='Fully Optimized')
ax.bar([p + 0.2 for p in x], half_optimized, width=0.2, label='Half Optimized')
ax.bar([p + 0.4 for p in x], no_optimization, width=0.2, label='No Optimization')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_ylabel('Time (microseconds)')
ax.set_title('Performance Comparison')
ax.set_xticks([p + 0.2 for p in x])
ax.set_xticklabels(categories)
ax.legend()

# Show the plot
plt.show()
```
