# Plots for Problem 1 and Problem 2 times 

```python
import matplotlib.pyplot as plt
```

```python
# Plots for problem 1 


# Data
categories = ['Total', 'Recursive', 'Reverse']
fully_optimized = [0.21, 0.04, 0.04]  # Fully optimization times
no_optimization = [3.58, 1.12, 2.33]  # No optimization times

# Create figure and axis
fig, ax = plt.subplots()

# Set the position of the bars on the x-axis
x = range(len(categories))

# Make the plot
ax.bar(x, fully_optimized, width=0.2, label='Fully Optimized')
ax.bar([p + 0.4 for p in x], no_optimization, width=0.2, label='No Optimization')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_ylabel('Time (microseconds)')
ax.set_title('Performance Comparison')
ax.set_xticks([p + 0.2 for p in x])
ax.set_xticklabels(categories)
ax.legend()

# Show the plot
plt.show()
```

```python
# Graph for problem 2

# Data preparation
functions = [
    'store_and_sort_bitsets_WoLoops',
    'store_and_sort_bitsets_loops'
]

functions_MOPS = [
    'population_count',
    'count_trailing_zeros',
    'popcount_Noloop',
    'countr_zero_Noloop',
    'count_occupied_orbitals_optimized',
    'std::popcount',
    'std::countr_zero'
]

# Times in microseconds
times_full_optimized = [7.58, 7.75]
times_no_optimization = [56.79, 56.58]

# Times in microseconds
times_full_optimized_MOPS = [0.04, 0.00, 0.04, 0.00, 0.04,0.04,0.04]
times_no_optimization_MOPS = [326918.29, 15089.50, 12162.75, 12909.83, 24661.79, 1895.79,2435.29]

iterations = 1_000_000  # 1 million iterations for all

# Correcting the MOPS calculation to convert time from microseconds to seconds before calculating MOPS
mops_full_optimized = [iterations / (time * 1e-6) if time != 0 else 0 for time in times_full_optimized_MOPS]
mops_no_optimization = [iterations / (time * 1e-6) if time != 0 else 0 for time in times_no_optimization_MOPS]

# Plotting the data again with corrected MOPS calculation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Time plot
ax1.barh(functions, times_no_optimization, label='No Optimization', color='red')
ax1.barh(functions, times_full_optimized, label='Fully Optimized', left=[sum(x) for x in zip(times_no_optimization)], color='green')
ax1.set_xlabel('Time (microseconds)')
ax1.set_title('Execution Time for Each Function')
ax1.legend()

# MOPS plot
ax2.barh(functions_MOPS, times_no_optimization_MOPS, label='No Optimization', color='red')
ax2.barh(functions_MOPS, times_full_optimized_MOPS, label='Fully Optimized', left=[sum(x) for x in zip(times_no_optimization_MOPS)], color='green')
ax2.set_xlabel('MOPS (Million Operations Per Second)')
ax2.set_title('Performance (MOPS) for Each Function')
ax2.legend()

plt.tight_layout()
plt.show()

# having a werid bug with the legend but it still runs 
()
```

```python
# Calculation of MOPS for fully optimiz

# Timing results in microseconds
population_count_time = 0.08  # microseconds
count_trailing_zeros_time = 0.04  # microseconds
popcount_Noloop_time = 0.04  # microseconds
countr_zero_Noloop_time = 0.04  # microseconds
countr_zero_std_time = 0.04  # microseconds
popcount_std_time = 0.04  # microseconds
iterations = 1000000  # Number of iterations

# Convert timing results to seconds
population_count_seconds = population_count_time * 1e-6
count_trailing_zeros_seconds = count_trailing_zeros_time * 1e-6
popcount_Noloop_seconds = popcount_Noloop_time * 1e-6
countr_zero_Noloop_seconds = countr_zero_Noloop_time * 1e-6
countr_zero_std_seconds = countr_zero_std_time * 1e-6
popcount_std_seconds = popcount_std_time * 1e-6

# Calculate iterations per second (IPS)
population_count_ips = iterations / population_count_seconds
count_trailing_zeros_ips = iterations / count_trailing_zeros_seconds
popcount_Noloop_ips = iterations / popcount_Noloop_seconds
countr_zero_Noloop_ips = iterations / countr_zero_Noloop_seconds
countr_zero_std_ips = iterations / countr_zero_std_seconds
popcount_std_ips = iterations / popcount_std_seconds

# Print results
print("Iterations per second (IPS):")
print("Total population_count: {:.2f}".format(population_count_ips))
print("Total count_trailing_zeros: {:.2f}".format(count_trailing_zeros_ips))
print("Total popcount_Noloop: {:.2f}".format(popcount_Noloop_ips))
print("Total countr_zero_Noloop: {:.2f}".format(countr_zero_Noloop_ips))
print("Total countr_zero from std: {:.2f}".format(countr_zero_std_ips))
print("Total popcount from std: {:.2f}".format(popcount_std_ips))
```
